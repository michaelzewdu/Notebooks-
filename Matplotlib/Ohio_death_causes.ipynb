```python
import matplotlib.pyplot as plt
```

```python

death_causes = 'Chronic', 'Infectious', 'Unintentional injury', 'Intentional injury', 'sucide'
percentile = [51, 20, 8, 2, 19]
type(death_causes)
```

```python
#Setting the figure size
plt.figure(figsize=(10,10))
#explode sets how much each pie slice distants itself from the center
explode = (0.08,0.1,0,0,0.075)

plt.pie(percentile, explode=explode, labels=death_causes, shadow= True, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
