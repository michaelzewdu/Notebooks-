```python
import seaborn as sns
%matplotlib inline
```

```python
flight_data = sns.load_dataset('flights')
flight_data.head()
```

```python
# Using pivot method the dataset can be transformed to a format suitable for a heatmap
flight_data = flight_data.pivot('year', 'month','passengers')
flight_data
```

```python
sns.heatmap(flight_data)
```
