```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
randNumber = np.random.rand(20)
randNumber
```

```python
# Selecting the style of plot
style.use('ggplot')

# Plot the reandom numbers
plt.plot(randNumber,'g',label='line one', linewidth=2)

plt.xlabel('Range')

plt.ylabel('Numbers')
#title of the plot
plt.title('Mike\'s first plot')
```

```python
plt.legend()
plt.show()
```
