```python
import pandas as pd
import seaborn as sns
```

```python
auto_dataset = pd.read_csv('../../Datasets/Auto/auto-mpg.csv')
auto_dataset.head()
```

```python
def orginRenamer(orgin):
    if(orgin==1):
        return 'USA'
    elif(orgin==2):
        return 'Europe'
    else:
        return 'Asia'

#Renaming the origin column using the orginRenamer function above
auto_dataset['origin']=auto_dataset['origin'].apply(orginRenamer)
auto_dataset.head()
```

```python
sns.pairplot(data=auto_dataset[['mpg', 'weight', 'origin']], hue= 'origin',size=4)
```
